```python
import cv2 

import uuid

import os

import time
```

# Define Images to Collect

```python
labels = ['Small Box', 'Medium Box', 'Paper Pouch']
number_imgs = 3
```

# Setting up Folders 

```python
IMAGES_PATH = os.path.join('Tensorflow', 'workspace', 'collectedimages')
```

```python
IMAGES_PATH
```

```python
if not os.path.exists(IMAGES_PATH):
    if os.name == 'posix':
        !mkdir -p {IMAGES_PATH}
    if os.name == 'nt':
         !mkdir {IMAGES_PATH}
for label in labels:
    path = os.path.join(IMAGES_PATH, label)
    if not os.path.exists(path):
        !mkdir {path}
```

# Capture Images

```python
for label in labels:
    cap = cv2.VideoCapture(0) #v.v.imp connects to webcam to capture device
    print('Collecting images for {}'.format(label))
    time.sleep(5) #timer to give time btw intervals
    for imgnum in range(number_imgs):
        print('Collecting image {}'.format(imgnum))
        ret, frame = cap.read()
        imgname = os.path.join(IMAGES_PATH,label,label+'.'+'{}.jpg'.format(str(uuid.uuid1()))) #create unique label
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame', frame)
        time.sleep(10) #take a break again

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
